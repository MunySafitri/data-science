```python
import numpy as np 
import pandas as pd 
```

```python
df_can = pd.read_csv('Angka Partisipasi Sekolah 2017 (1).csv')
```

```python
df_can.head(34)
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

print('Matplotlib version: ', mpl.__version__)

print(plt.style.available)
mpl.style.use(['ggplot'])
```

```python
#membuat visualisasi plot stacked
df_plot_stacked = df_can.copy()
df_plot_stacked["Rata-rata"] = (df_plot_stacked["SD"] + df_plot_stacked["SMP"] + df_plot_stacked["SMA"] + df_plot_stacked["PT"])/4
df_plot_stacked["Total"] = df_plot_stacked["SD"] + df_plot_stacked["SMP"] + df_plot_stacked["SMA"] + df_plot_stacked["PT"]
df_plot_stacked["SD"] = df_plot_stacked["SD"]/df_plot_stacked["Total"]*df_plot_stacked["Rata-rata"]
df_plot_stacked["SMP"] = df_plot_stacked["SMP"]/df_plot_stacked["Total"]*df_plot_stacked["Rata-rata"]
df_plot_stacked["SMA"] = df_plot_stacked["SMA"]/df_plot_stacked["Total"]*df_plot_stacked["Rata-rata"]
df_plot_stacked["PT"] = df_plot_stacked["PT"]/df_plot_stacked["Total"]*df_plot_stacked["Rata-rata"]

df_plot_stacked = df_plot_stacked.sort_values('Rata-rata',ascending=False)
df_plot_stacked.drop('Rata-rata', axis=1, inplace=True)
df_plot_stacked.drop('Total', axis=1, inplace=True)
```

```python
df_plot_stacked.plot(x='Provinsi', kind='bar', 
                     stacked=True, title='Persentase Partisipasi Pendidikan Tiap Provinsi dari Terbesar ke Terkecil Tahun 2017\n',
                     figsize=(20, 8), yticks=range(0,101,5), ylim=(0,100))
```

```python
#membuat visualisasi bar chart
df_bar = df_can.copy()
df_bar.set_index('Provinsi', inplace=True)
df_bar = df_bar.sort_values('PT',ascending=False)
```

```python
import seaborn as sns
# Set the width and height of the figure
plt.figure(figsize=(6,10))

# Add title
plt.title("Persentase Masyarakat yang Menempuh Pendidikan Perguruan Tinggi \ndari Terbesar ke Terkecil di Tiap Provinsi Tahun 2017\n")

# sns.barplot(x=df_bar.index, y=df_bar['SMP'], order=df_bar_pt['SMP'].index[0:10])
sns.barplot(x=df_bar['PT'], y=df_bar.index)

plt.xlim(0,100)
# Add label for vertical axis
plt.xlabel("Persentase")
```

```python
df_bar = df_bar.sort_values('SMP',ascending=False)

# Set the width and height of the figure
plt.figure(figsize=(6,10))

# Add title
plt.title("Persentase Masyarakat yang Menempuh Pendidikan SMP dari Terbesar ke Terkecil di Tiap Provinsi Tahun 2017\n")

# sns.barplot(x=df_bar_pt.index, y=df_bar_pt['SMP'], order=df_bar_pt['SMP'].index[0:10])
sns.barplot(x=df_bar['SMP'], y=df_bar.index)

plt.xlim(75,100)
# Add label for vertical axis
plt.xlabel("Persentase")
```

```python
df_bar = df_bar.sort_values('SMA',ascending=False)

# Set the width and height of the figure
plt.figure(figsize=(6,10))

# Add title
plt.title("Persentase Masyarakat yang Menempuh Pendidikan SMA dari Terbesar ke Terkecil di Tiap Provinsi Tahun 2017\n")

# sns.barplot(x=df_bar_pt.index, y=df_bar_pt['SMP'], order=df_bar_pt['SMP'].index[0:10])
sns.barplot(x=df_bar['SMA'], y=df_bar.index)

plt.xlim(50,100)
# Add label for vertical axis
plt.xlabel("Persentase")
```

```python
df_bar = df_bar.sort_values('SD',ascending=False)

# Set the width and height of the figure
plt.figure(figsize=(6,10))

# Add title
plt.title("Persentase Masyarakat yang Menempuh Pendidikan SD dari Terbesar ke Terkecil di Tiap Provinsi Tahun 2017\n")

# sns.barplot(x=df_bar.index, y=df_bar['SMP'], order=df_bar['SMP'].index[0:10])
sns.barplot(x=df_bar['SD'], y=df_bar.index)

plt.xlim(80,100)
# Add label for vertical axis
plt.xlabel("Persentase")
```

```python
#membuat histogram

plt.figure(figsize=(15,6))
df_plot_histogram = df_can.copy()

sns.distplot(a=df_plot_histogram['SD'], label="SD", kde=False)
sns.distplot(a=df_plot_histogram['SMP'], label="SMP", kde=False)
sns.distplot(a=df_plot_histogram['SMA'], label="SMA", kde=False)
sns.distplot(a=df_plot_histogram['PT'], label="PT", kde=False)

# Add title
plt.title("Distribusi Persentase Angka Partisipasi Sekolah Tahun 2017")
plt.xlabel("Persentase Partisipasi")
plt.ylabel("Banyak Data")


# Force legend to appear
plt.legend()
```

```python
#membuat kde_plot
plt.figure(figsize=(15,6))

sns.kdeplot(data=df_plot_histogram['SD'], label="SD", shade=True)
sns.kdeplot(data=df_plot_histogram['SMP'], label="SMP", shade=True)
sns.kdeplot(data=df_plot_histogram['SMA'], label="SMA", shade=True)
sns.kdeplot(data=df_plot_histogram['PT'], label="PT", shade=True)

# Add title
plt.title("Distribusi Persentase Angka Partisipasi Sekolah Tahun 2017")
plt.xlabel("Persentase Partisipasi")
plt.ylabel("Banyak Data")


# Force legend to appear
plt.legend()
```

```python
df_heatmap_2017 = pd.DataFrame({
    "SD":df_can["SD"], 
    "SMP":df_can["SMP"], 
    "SMA":df_can["SMA"], 
    "PT":df_can["PT"],
    "Provinsi" :df_can["Provinsi"]
})
df_heatmap_2017.set_index("Provinsi", inplace=True)
```

```python
# Add title
plt.figure(figsize=(10,10))

plt.title("Persentase Partisipasi Sekolah Tahun  2017, Berdasarkan Provinsi\n")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=df_heatmap_2017, annot=True)

# Add label for horizontal axis
plt.xlabel("Partisipasi Sekolah")
```
